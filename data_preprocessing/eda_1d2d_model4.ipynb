### Post-processing verification

Run **after** the dataset has been processed above. Validates the saved artifacts (`constant_values.npz`, per-event `dynamic_values_event_*.npz`) and a few constructed samples for: feature/element/timestep shape consistency, NaN/Inf, edge-index bounds, 1D node-mapping bijectivity, timestep cadence & truncation bookkeeping, and boundary masks. Raises `AssertionError` if any hard **ERROR** is found; **WARNINGS** are sanity flags to eyeball.

```python
import numpy as np
import pandas as pd
import torch


def run_postprocessing_checks(dataset, sample_indices=None, verbose=True):
    """Validate the outputs of FloodEvent1D2DDataset AFTER processing.

    Reads the artifacts written by `process()` (constant_values.npz and the
    per-event dynamic_values_event_*.npz files) plus a few constructed samples
    from `dataset.get()`, and checks them for structural correctness, missing
    values, and truncation/downsampling consistency.

    Two severities are collected:
      * ERROR   - must be zero for a clean, open-sourceable dataset. An
                  AssertionError is raised at the end if any are present.
      * WARNING - sanity flags worth eyeballing (not necessarily wrong).

    Returns a dict with the collected messages and summary info.
    """
    errors, warnings, info = [], [], {}

    def err(msg):
        errors.append(msg)

    def warn(msg):
        warnings.append(msg)

    # ------------------------------------------------------------------ #
    # 1. Load processed constant artifacts                                #
    # ------------------------------------------------------------------ #
    const = np.load(dataset.processed_paths[3], allow_pickle=True)
    edge_index = const["edge_index"]
    static_nodes = const["static_nodes"]
    static_edges = const["static_edges"]
    edge_index_1d = const["edge_index_1d"]
    edge_index_1d2d = const["edge_index_1d_2d"]
    static_nodes_1d = const["static_nodes_1d"]
    static_edges_1d = const["static_edges_1d"]

    num_2d_nodes = static_nodes.shape[0]
    num_2d_edges = static_edges.shape[0]
    num_1d_nodes = static_nodes_1d.shape[0]
    num_1d_edges = static_edges_1d.shape[0]
    info.update(
        num_2d_nodes=num_2d_nodes,
        num_2d_edges=num_2d_edges,
        num_1d_nodes=num_1d_nodes,
        num_1d_edges=num_1d_edges,
        num_1d2d_connections=int(edge_index_1d2d.shape[1]),
        num_events=len(dataset.hec_ras_run_ids),
        previous_timesteps=dataset.previous_timesteps,
        timestep_interval=dataset.timestep_interval,
        spin_up_time=dataset.spin_up_time,
        time_from_peak=dataset.time_from_peak,
        normalized=dataset.is_normalized,
    )

    # ------------------------------------------------------------------ #
    # 2. Static feature-matrix widths match the declared feature lists    #
    # ------------------------------------------------------------------ #
    width_checks = [
        ("2D static nodes", static_nodes.shape[1], dataset.STATIC_NODE_FEATURES),
        ("2D static edges", static_edges.shape[1], dataset.STATIC_EDGE_FEATURES),
        ("1D static nodes", static_nodes_1d.shape[1], dataset.STATIC_1D_NODE_FEATURES),
        ("1D static edges", static_edges_1d.shape[1], dataset.STATIC_1D_EDGE_FEATURES),
    ]
    for name, got, feats in width_checks:
        if got != len(feats):
            err(f"{name}: {got} columns but {len(feats)} feature names declared {feats}.")

    # edge_index width must match the static edge count
    if edge_index.shape[1] != num_2d_edges:
        err(f"2D edge_index has {edge_index.shape[1]} edges but static_edges has {num_2d_edges}.")
    if edge_index_1d.shape[1] != num_1d_edges:
        err(f"1D edge_index has {edge_index_1d.shape[1]} edges but static_edges_1d has {num_1d_edges}.")

    # ------------------------------------------------------------------ #
    # 3. Edge-index integrity (shape + node-id bounds)                    #
    # ------------------------------------------------------------------ #
    def check_ei(ei, n_nodes, label):
        if ei.shape[0] != 2:
            err(f"{label} edge_index shape {ei.shape}, expected (2, E).")
            return
        if ei.size and (ei.min() < 0 or ei.max() >= n_nodes):
            err(f"{label} edge_index references node id outside [0,{n_nodes - 1}] "
                f"(min={int(ei.min())}, max={int(ei.max())}).")

    check_ei(edge_index, num_2d_nodes, "2D")
    check_ei(edge_index_1d, num_1d_nodes, "1D")

    if edge_index_1d2d.shape[0] != 2:
        err(f"1D-2D edge_index shape {edge_index_1d2d.shape}, expected (2, C).")
    else:
        c1d, c2d = edge_index_1d2d[0], edge_index_1d2d[1]
        if c1d.size and (c1d.min() < 0 or c1d.max() >= num_1d_nodes):
            err(f"1D-2D: 1D node id out of range [0,{num_1d_nodes - 1}] "
                f"(min={int(c1d.min())}, max={int(c1d.max())}). "
                f"Confirm node_1d column aligns with Nodes_1D fid ordering.")
        if c2d.size and (c2d.min() < 0 or c2d.max() >= num_2d_nodes):
            err(f"1D-2D: 2D node id out of range [0,{num_2d_nodes - 1}] "
                f"(min={int(c2d.min())}, max={int(c2d.max())}). "
                f"2D ids must match indexing AFTER ghost-node removal.")

    # ------------------------------------------------------------------ #
    # 4. 1D node mapping must be a valid bijection onto all 1D nodes      #
    # ------------------------------------------------------------------ #
    mapping = dict(dataset.node_1d_mapping or {})
    if mapping:
        keys = sorted(int(k) for k in mapping.keys())
        vals = [int(v) for v in mapping.values()]
        if keys != list(range(num_1d_nodes)):
            err(f"node_1d_mapping keys are not exactly 0..{num_1d_nodes - 1} "
                f"({len(keys)} keys for {num_1d_nodes} 1D nodes). Unmapped fids get "
                f"uninitialised np.empty() garbage in the dynamic 1D node array.")
        if len(set(vals)) != len(vals):
            err("node_1d_mapping has duplicate target columns -> a HEC-RAS column is reused.")
        if vals and min(vals) < 0:
            err("node_1d_mapping has negative HEC-RAS column indices.")
    else:
        warn("node_1d_mapping is empty; 1D dynamic node ordering assumed identical to HEC-RAS output.")

    # ------------------------------------------------------------------ #
    # 5. NaN / Inf in static arrays                                       #
    # ------------------------------------------------------------------ #
    static_arrays = [
        ("static_nodes_2d", static_nodes),
        ("static_edges_2d", static_edges),
        ("static_nodes_1d", static_nodes_1d),
        ("static_edges_1d", static_edges_1d),
    ]
    for name, arr in static_arrays:
        n_nan, n_inf = int(np.isnan(arr).sum()), int(np.isinf(arr).sum())
        if n_nan:
            err(f"{name} contains {n_nan} NaN values.")
        if n_inf:
            err(f"{name} contains {n_inf} Inf values.")

    # light physical sanity on a few static features (warnings only)
    def col(feats, arr, name):
        return arr[:, feats.index(name)] if name in feats else None

    area = col(dataset.STATIC_NODE_FEATURES, static_nodes, "area")
    if area is not None and (area <= 0).any():
        warn(f"2D node 'area' has {int((area <= 0).sum())} non-positive values.")
    rough = col(dataset.STATIC_NODE_FEATURES, static_nodes, "roughness")
    if rough is not None and ((rough <= 0).any() or (rough > 1).any()):
        warn("2D node 'roughness' (Manning n) outside the usual (0, 1] range.")
    shape_1d = col(dataset.STATIC_1D_EDGE_FEATURES, static_edges_1d, "shape")
    if shape_1d is not None and not np.isin(np.unique(shape_1d), [0, 1, 2, 3]).all():
        warn(f"1D edge 'shape' has unexpected encodings: {np.unique(shape_1d)} (expected subset of 0-3).")
    diam = col(dataset.STATIC_1D_EDGE_FEATURES, static_edges_1d, "diameter")
    if diam is not None and (diam <= 0).any():
        warn(f"1D edge 'diameter' has {int((diam <= 0).sum())} non-positive values.")

    # ------------------------------------------------------------------ #
    # 6. Per-event dynamic arrays: shapes, NaN/Inf, truncation, cadence   #
    # ------------------------------------------------------------------ #
    total_expected_rollout = 0
    dyn_feat_dims = {
        "dynamic_nodes": (num_2d_nodes, len(dataset.DYNAMIC_NODE_FEATURES)),
        "dynamic_edges": (num_2d_edges, len(dataset.DYNAMIC_EDGE_FEATURES)),
        "dynamic_nodes_1d": (num_1d_nodes, len(dataset.DYNAMIC_1D_NODE_FEATURES)),
        "dynamic_edges_1d": (num_1d_edges, len(dataset.DYNAMIC_1D_EDGE_FEATURES)),
    }
    per_event = []
    for i, run_id in enumerate(dataset.hec_ras_run_ids):
        dyn = np.load(dataset.processed_paths[i + 4], allow_pickle=True)
        T = dyn["dynamic_nodes"].shape[0]
        per_event.append((run_id, T))

        for key, (n_elem, n_feat) in dyn_feat_dims.items():
            a = dyn[key]
            if a.shape[0] != T:
                err(f"event {run_id}: '{key}' has {a.shape[0]} timesteps, expected {T}.")
            if a.shape[1] != n_elem:
                err(f"event {run_id}: '{key}' has {a.shape[1]} elements, expected {n_elem} (static mismatch).")
            if a.shape[2] != n_feat:
                err(f"event {run_id}: '{key}' has {a.shape[2]} features, expected {n_feat}.")
            if np.isnan(a).any():
                err(f"event {run_id}: '{key}' contains {int(np.isnan(a).sum())} NaN values.")
            if np.isinf(a).any():
                err(f"event {run_id}: '{key}' contains {int(np.isinf(a).sum())} Inf values.")

        # water_level should actually vary -- a flat series usually means the
        # wrong perimeter_name / network_name was read.
        wl = dyn["dynamic_nodes"][:, :, dataset.DYNAMIC_NODE_FEATURES.index("water_level")]
        if np.allclose(wl, wl.flat[0]):
            warn(f"event {run_id}: 2D water_level is constant across all cells/timesteps -- check perimeter_name.")
        wl1d = dyn["dynamic_nodes_1d"][:, :, dataset.DYNAMIC_1D_NODE_FEATURES.index("water_level")]
        if np.allclose(wl1d, wl1d.flat[0]):
            warn(f"event {run_id}: 1D water_level is constant -- check network_name / node_1d_mapping.")

        # timestamp cadence: strictly increasing and equal to timestep_interval
        ts = dyn["event_timesteps"]
        if len(ts) != T:
            err(f"event {run_id}: event_timesteps length {len(ts)} != {T} dynamic timesteps.")
        if len(ts) >= 2:
            tstamps = pd.to_datetime([str(t) for t in ts])
            diffs = np.diff(tstamps.astype("int64").to_numpy()) / 1e9  # seconds
            if (diffs <= 0).any():
                err(f"event {run_id}: event_timesteps are not strictly increasing.")
            elif not np.allclose(diffs, dataset.timestep_interval, atol=1e-6):
                warn(f"event {run_id}: timestep spacing {sorted(set(np.round(diffs, 3)))}s "
                     f"differs from configured timestep_interval {dataset.timestep_interval}s.")

        # rollout accounting: first `previous_timesteps` are input-only, last is label
        event_rollout = T - dataset.previous_timesteps - 1
        if event_rollout <= 0:
            err(f"event {run_id}: {T} timesteps give non-positive rollout length "
                f"({event_rollout}); event is too short for previous_timesteps={dataset.previous_timesteps}.")
        total_expected_rollout += max(event_rollout, 0)

    info["per_event_timesteps"] = per_event

    # ------------------------------------------------------------------ #
    # 7. Global index bookkeeping                                         #
    # ------------------------------------------------------------------ #
    if total_expected_rollout != dataset.total_rollout_timesteps:
        err(f"Sum of per-event rollout ({total_expected_rollout}) != "
            f"dataset.total_rollout_timesteps ({dataset.total_rollout_timesteps}).")
    if len(dataset) != dataset.total_rollout_timesteps:
        err(f"len(dataset)={len(dataset)} != total_rollout_timesteps={dataset.total_rollout_timesteps}.")
    if list(dataset.event_start_idx) != sorted(dataset.event_start_idx):
        err(f"event_start_idx is not monotonically increasing: {dataset.event_start_idx}.")
    if len(dataset.event_start_idx) != len(dataset.hec_ras_run_ids):
        err("event_start_idx length does not match the number of events.")

    # ------------------------------------------------------------------ #
    # 8. Downsampling divisibility (needs process()-time attributes)      #
    # ------------------------------------------------------------------ #
    base_intervals = getattr(dataset, "_event_base_timestep_interval", None)
    if base_intervals:
        for i, run_id in enumerate(dataset.hec_ras_run_ids):
            base = base_intervals[i]
            if base and dataset.timestep_interval % base != 0:
                err(f"event {run_id}: timestep_interval {dataset.timestep_interval}s is not a "
                    f"multiple of the raw interval {base}s -> downsampling is ill-defined.")

    # ------------------------------------------------------------------ #
    # 9. Boundary-condition masks                                         #
    # ------------------------------------------------------------------ #
    bc = getattr(dataset, "boundary_condition", None)
    if bc is not None and getattr(bc, "boundary_nodes_mask", None) is not None:
        if bc.boundary_nodes_mask.shape[0] != num_2d_nodes:
            err(f"boundary_nodes_mask length {bc.boundary_nodes_mask.shape[0]} != num_2d_nodes {num_2d_nodes}.")
        for mname in ("boundary_edges_mask", "inflow_edges_mask", "outflow_edges_mask"):
            m = getattr(bc, mname, None)
            if m is not None and m.shape[0] != num_2d_edges:
                err(f"{mname} length {m.shape[0]} != num_2d_edges {num_2d_edges}.")

    # ------------------------------------------------------------------ #
    # 10. Spot-check constructed samples from get()                       #
    # ------------------------------------------------------------------ #
    n = len(dataset)
    if sample_indices is None:
        sample_indices = sorted({0, n // 2, n - 1}) if n > 0 else []
    p1 = dataset.previous_timesteps + 1
    exp = {
        "x": (num_2d_nodes, len(dataset.STATIC_NODE_FEATURES) + len(dataset.DYNAMIC_NODE_FEATURES) * p1),
        "edge_attr": (num_2d_edges, len(dataset.STATIC_EDGE_FEATURES) + len(dataset.DYNAMIC_EDGE_FEATURES) * p1),
        "x_1d": (num_1d_nodes, len(dataset.STATIC_1D_NODE_FEATURES) + len(dataset.DYNAMIC_1D_NODE_FEATURES) * p1),
        "edge_attr_1d": (num_1d_edges, len(dataset.STATIC_1D_EDGE_FEATURES) + len(dataset.DYNAMIC_1D_EDGE_FEATURES) * p1),
        "y": (num_2d_nodes, 1),
        "y_edge": (num_2d_edges, 1),
        "y_1d": (num_1d_nodes, 1),
        "y_1d_edge": (num_1d_edges, 1),
    }
    for idx in sample_indices:
        try:
            d = dataset.get(idx)
        except Exception as e:  # noqa: BLE001
            err(f"dataset.get({idx}) raised {type(e).__name__}: {e}")
            continue
        for attr, shape in exp.items():
            t = getattr(d, attr, None)
            if t is None:
                err(f"sample {idx}: missing attribute '{attr}'.")
                continue
            if tuple(t.shape) != shape:
                err(f"sample {idx}: '{attr}' shape {tuple(t.shape)}, expected {shape}.")
            if not torch.isfinite(t).all():
                err(f"sample {idx}: '{attr}' contains non-finite values.")

    # ------------------------------------------------------------------ #
    # Report                                                              #
    # ------------------------------------------------------------------ #
    if verbose:
        print("=" * 70)
        print(f"POST-PROCESSING CHECK  |  mode={dataset.mode}  |  events={info['num_events']}")
        print("=" * 70)
        print(f"  2D: {num_2d_nodes:>7} nodes | {num_2d_edges:>7} edges")
        print(f"  1D: {num_1d_nodes:>7} nodes | {num_1d_edges:>7} edges | "
              f"{info['num_1d2d_connections']} 1D-2D links")
        print(f"  rollout timesteps = {dataset.total_rollout_timesteps} | "
              f"previous_timesteps = {dataset.previous_timesteps} | "
              f"dt = {dataset.timestep_interval}s | normalized = {dataset.is_normalized}")
        print(f"  spin_up_time = {dataset.spin_up_time} | time_from_peak = {dataset.time_from_peak}")
        print("-" * 70)
        if warnings:
            print(f"WARNINGS ({len(warnings)}):")
            for w in warnings:
                print(f"  [warn] {w}")
        else:
            print("WARNINGS: none")
        print("-" * 70)
        if errors:
            print(f"ERRORS ({len(errors)}):")
            for e in errors:
                print(f"  [ERROR] {e}")
        else:
            print("ERRORS: none  ->  dataset passed all structural checks.")
        print("=" * 70)

    info["errors"] = errors
    info["warnings"] = warnings
    assert not errors, f"{len(errors)} post-processing check(s) failed - see output above."
    return info
```
